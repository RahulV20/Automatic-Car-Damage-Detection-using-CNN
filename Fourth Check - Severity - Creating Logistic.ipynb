```python
# imports required packages

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
import numpy as np
import h5py
import os
import json
import pickle
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
config={
  "model"           : "vgg16",
  "weights"         : "imagenet",

  "features_path"   : "/Users/RAHUL/Downloads/car-damage-check3/features.h5",
  "labels_path"     : "/Users/RAHUL/Downloads/car-damage-check3/labels.h5",
  "classifier_path" : "/Users/RAHUL/Downloads/car-damage-check3/classifier.pickle",
  "model_path"      : "/Users/RAHUL/Downloads/car-damage-check3/model",

  "test_size"       : 0.20,
  "seed"            : 9,
}
```

```python
# config variables
test_size     = config["test_size"]
seed      = config["seed"]
features_path   = config["features_path"]
labels_path   = config["labels_path"]
classifier_path = config["classifier_path"]
```

```python
# import features and labels
h5f_data  = h5py.File(features_path, 'r')
h5f_label = h5py.File(labels_path, 'r')

features_string = h5f_data['dataset_1']
labels_string   = h5f_label['dataset_1']

features = np.array(features_string)
labels   = np.array(labels_string)

h5f_data.close()
h5f_label.close()
```

```python
# verify the shape of features and labels
print ("[INFO] features shape: {}".format(features.shape))
print ("[INFO] labels shape: {}".format(labels.shape))

print ("[INFO] training started...")
# split the training and testing data
(trainData, testData, trainLabels, testLabels) = train_test_split(np.array(features),
                                                                  np.array(labels),
                                                                  test_size=test_size,
                                                                  random_state=seed)

print ("[INFO] splitted train and test data...")
print ("[INFO] train data  : {}".format(trainData.shape))
print ("[INFO] test data   : {}".format(testData.shape))
print ("[INFO] train labels: {}".format(trainLabels.shape))
print ("[INFO] test labels : {}".format(testLabels.shape))
```

```python
# use logistic regression as the model
print ("[INFO] creating model...")
model = LogisticRegression(random_state=seed)
model.fit(trainData, trainLabels)
```

```python
# evaluate the model of test data
preds = model.predict(testData)
```

```python
preds.shape
```

```python
# dump classifier to file
print ("[INFO] saving model...")
pickle.dump(model, open(classifier_path, 'wb'))
```

```python
# plot the confusion matrix
cm = confusion_matrix(testLabels, preds)
```
